```python
## Do the distributions for different numerical variables look like a normal distribution

# '''''''''''''''''''''''''''''''''''
# THIS IS ACTUALLY NOT IMPORTANT HERE; BUT WE´LL SEE IF IT HELPS OR HURTS
# ''''''''''''''''''''''''''''''''''

def log_claim_amount(x):    # define a function to use it to find a ND
    x = np.log2(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN # We are returning NaNs so that we can replace them with means later
    
numerical['log_claim_amount'] = list(map(log_claim_amount, numerical['total claim amount']))
# Now we replace the NAN by the mean
numerical['log_claim_amount'] = numerical['log_claim_amount'].fillna(np.mean(numerical['log_claim_amount']))
sns.displot(numerical['log_claim_amount'])
plt.show()
```

```python

def log_months_policy(x):
    x = np.log2(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN # We are returning NaNs so that we can replace them with means later
    
numerical['log_months_policy'] = list(map(log_months_policy, numerical['months since policy inception']))
# Now we replace the NAN by the mean
numerical['log_months_policy'] = numerical['log_months_policy'].fillna(np.mean(numerical['log_months_policy']))
sns.displot(numerical['log_months_policy'])
plt.show()
```

```python
# For the numerical variables, check the multicollinearity between the features.
#Please note that we will use the column total_claim_amount later as the target variable.
fig, ax=plt.subplots(figsize=(10,9))
#make graphs fit
# you can make your graphs look nice with axis (look it up)

correlations_matrix = numerical.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
# I will do OneHotEnconder for the rest categorical columns

from sklearn.preprocessing import OneHotEncoder

# drop coverage & vehicle size ##########
X_cat_hot = X_cat.drop(['vehicle size', 'coverage', 'customer'], axis=1)
encoderhot = OneHotEncoder(drop='first').fit(X_cat_hot)

cols=encoderhot.get_feature_names_out(input_features=X_cat_hot.columns)   # takes the column names from CAT

encodedhot = encoderhot.transform(X_cat_hot).toarray()  # array: display or range of a particular type of thing
onehot_encoded = pd.DataFrame(encodedhot,columns=cols)
onehot_encoded

############## option internet  - check how it works later

# transformed_data = onehotencoder.fit_transform(data[categorical_cols])
# # the above transformed_data is an array so convert it to dataframe
# encoded_data = pd.DataFrame(transformed_data, index=data.index)
```

```python
from sklearn.preprocessing import LabelEncoder

print(X_cat['coverage'].unique())

print(X_cat['vehicle size'].unique())
```

```python
# Create a list with the values for the label encoder
#le = preprocessing.LabelEncoder()

# just for vehicle size
X_cat_label_v = X_cat.drop(['customer', 'state', 'gender', 'response', 'education',
       'employment status', 'location code', 'marital status', 'policy type',
       'policy', 'renew offer type', 'sales channel', 'vehicle class', 'coverage'], axis = 1)

# just for coverage
X_cat_label_c = X_cat.drop(['customer', 'state', 'gender', 'response', 'education',
       'employment status', 'location code', 'marital status', 'policy type',
       'policy', 'renew offer type', 'sales channel', 'vehicle class', 'vehicle size'], axis = 1)

# just for vehicle size
label_encoded_v = LabelEncoder().fit(['Small', 'Medsize', 'Large']).transform(X_cat_label_v) # give a value to each type
label_encoded_v = pd.DataFrame(label_encoded_v,columns=X_cat_label_v.columns)
display(label_encoded_v.head())

# just for coverage
label_encoded_c = LabelEncoder().fit(['Basic', 'Premium', 'Extended']).transform(X_cat_label_c) # ordered wrt value counts
label_encoded_c = pd.DataFrame(label_encoded_c,columns=X_cat_label_c.columns)
display(label_encoded_c.head())

#If order is irrelevant then inside .fit() could just have the list name; however, in that case, better use onehotencoder?

# # don´t think I need this
#label_encoded['gender'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)    # .2 = 20% of the data will be used for testing
# split the data in Test Set (X,y) & Train Set (X,y)
```

```python
# Apply linear regression.
```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

### Model Validation

```python
# R2
from sklearn.metrics import r2_score

predictions = lm.predict(X_train)
r2_score(y_train, predictions)
# This tells us what is currently happening 'training'
```

```python
# Now we evaluate the test

predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

# Improve the Linear Regression Model

```python
#dropped month as it is probably irrelevant for the analysis

def improve():
    data_normalized1 = data_normalized.drop(['month'],axis=1)
        
    X1 = pd.concat([data_normalized1,label_encoded_c, label_encoded_v, onehot_encoded], axis=1)
    X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y, test_size=0.2)
    lm1 = linear_model.LinearRegression()
    lm1.fit(X1_train,y1_train)
    predictions1 = lm1.predict(X1_train)
    r2_score(y1_train, predictions1)
    predictions_test1 = lm1.predict(X1_test)
    print('y1 test: ',r2_score(y1_test, predictions_test1))
    mse1=mean_squared_error(y1_test,predictions_test1)
    print('mse: ',mse1)
    rmse1 = np.sqrt(mean_squared_error(y1_test,predictions_test1))
    print('rmse: ',rmse1)
    mae1 = mean_absolute_error(y1_test, predictions_test1)
    print('mae: ',mae1)
improve()
```

```python
# Original Data as comparison
# it improved by .15 by taking the month off.
print('Original Data y test: ', r2_score(y_test, predictions_test))
print('Original Data mse: ',mse)
print('Original Data mse: ',rmse)
print('Original Data mse: ',mae)
```
